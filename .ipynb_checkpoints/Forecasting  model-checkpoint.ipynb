```python
import torch
import torch.nn as nn
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score, precision_score
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('Occupancy_Estimation.csv')
columns_names = list(data.columns)
data
```

```python
#Feature selection
selected_data = data[['S1_Temp', 'S3_Temp', 'S5_CO2', 'S1_Light','S3_Light','S7_PIR', 'Room_Occupancy_Count']]
selected_data
```

```python
#Splitting data to train and test samples
X, y = selected_data[['S1_Temp', 'S3_Temp', 'S5_CO2', 'S1_Light','S3_Light','S7_PIR']], selected_data['Room_Occupancy_Count']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.75, shuffle = False, random_state = 42)
X_train
```

```python
#Scaling data by minmaxscaler with range = [-1, 1]
min_max_scaler = preprocessing.MinMaxScaler(feature_range = (-1, 1)) 
x_data_minmax_scaler = min_max_scaler.fit(X_train)
X_test_scaled = x_data_minmax_scaler.transform(X_test)
X_train_scaled = x_data_minmax_scaler.transform(X_train)
```

```python
#Building and training SVM model for classification to 4 classes (0, 1, 2, 3 humans in the room).
classifier_model_rbfkernel = svm.SVC(decision_function_shape ='ovo')
classifier_model_rbfkernel.fit(X_train_scaled, y_train)

classifier_model_linkernel = svm.SVC(decision_function_shape ='ovo', kernel = 'linear')
classifier_model_linkernel.fit(X_train_scaled, y_train)
```

```python
#Checking the results of the pridiction on test data 
y_test_pred1 = classifier_model_linkernel.predict(X_test_scaled)
y_test_pred2 = classifier_model_rbfkernel.predict(X_test_scaled)
indexes = np.array([i for i in range(len(y_test))])
plt.plot(indexes, y_test_pred1, indexes, y_test)
plt.legend(['y_test_pred1', 'y_test'])
plt.title('Prediction with linear kernel function')
```

```python
plt.plot(indexes, y_test_pred2, indexes, y_test)
plt.legend(['y_test_pred2', 'y_test'])
plt.title('Prediction with rbf kernel function')
```

```python
print(f'Prediction accuracy using SVM with linear kernel: {accuracy_score(y_test, y_test_pred1)}')
print(f'Prediction accuracy using SVM with rbf kernel: {accuracy_score(y_test, y_test_pred2)}')
```

```python
precision_score(y_test, y_test_pred1, average='macro')
precision_score(y_test, y_test_pred2, average='macro')
```

```python
train_data, test_data = data[:int(len(data)*0.75)], data[int(len(data)*0.75):]
X_train, X_test = train_data[columns_names[2:len(columns_names)-1]], test_data[columns_names[2:len(columns_names)-1]]
Y_train, Y_test = train_data[columns_names[len(columns_names)-1]], test_data[columns_names[len(columns_names)-1]] 
X_train
```

```python
def split_sequences(sequences, n_steps):    
    X, y = [],[]
    for i in range(len(sequences[0])):
        end_ix = i + n_steps
        if end_ix > len(sequences[0]):
            break
        
        X.append(sequences[0].iloc[i:end_ix, : ])
        y.append(sequences[1].iloc[end_ix-1])
        
    return torch.tensor(np.array(X)), torch.tensor(np.array(y))

X,Y = split_sequences((X_train, Y_train), 3) 
tuple(X[0].size())
```

```python
#define hyperparameters
batchsize = 100
#sequence_len = X.size()[1]
input_len = tuple(X[0].size())
hidden_size = 50
num_layers = 2
num_epochs = 5
learning_rate = 0.001
tuple(X[0].size())
```

```python
#create a model
class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(LSTM, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first = True)
        self.linear = nn.Linear(hidden_size, output_size)
        
    def forward(self, x, hidden = None):
        '''
        if hidden == None:
            self.hidden = (torch.zeros(1,1,self.hidden_size), torch.zeros(1,1,self.hidden_size))
        else:
            self.hidden = hidden
            
        lstm_out, self.hidden = self.lstm(x.view(len(x),1,-1), self.hidden)
        predictions = self.linear(lstm_out.view(len(x), -1))
        return predictions[-1], self.hidden
        '''
        h0 = torch.zeros(self.num_layers, self.hidden_size, dtype = torch.float64).to(device) 
        c0 = torch.zeros(self.num_layers, self.hidden_size, dtype = torch.float64).to(device)
        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])
        return out
        
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = LSTM(input_size = input_len[1], hidden_size = 50, num_layers = num_layers, output_size = 1).to(device)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr = learning_rate)

epochs = 600
model.train()
for epoch in range(epochs + 1):
    for x,y in zip(X, Y):
        print(x.size(0))
        y_pred, _ = model(x, None)
        optimizer.zero_grad()
        loss = criterion(y_pred, y)
        loss.backward()
        optimizer.step()
        
    if epoch % 10 == 0:
        print(f'epoch: {epoch:4} loss:{loss.item():10.8f}')
```

```python
X2, Y2 = split_sequences((X_train, Y_train), 1)
X2.size()[2]
X2[0]
```

```python
'''
Попытка сделать сделать обычную полносвязную нейронную сеть
'''
class NeuralNetwork(nn.Module):
    def __init__(self, sample_size):
        super().__init__()
        self.flatten = nn.Flatten()
        self.sample_size = sample_size
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(sample_size, 8),
            nn.ReLU(),
            nn.Linear(8, 4),
            nn.ReLU(),
            nn.Linear(4, 1),
        )

    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using {device} device")
model = NeuralNetwork(X2.size()[2]).to(device)
print(model)
```
